# Likelihood

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 10, 5000)
y1 = np.exp(-0.1*x**2)
y2 = np.exp(-0.1*(x-5)**2)
```

```python
plt.plot(x, y1, c="orange")
plt.plot(x, y2, c="b")
plt.show()
```

# 예시

```python
sample = [1, 4, 5, 6, 9]
```

```python
np.mean(sample)
```

```python
np.std(sample)**2
```

```python
mean = np.mean(sample)
sigma_square = np.std(sample)**2
x = np.linspace(-10, 10, 1000)
f_x = 1/(np.sqrt(2*np.pi*sigma_square))*np.exp(-(x-mean)**2/(2*sigma_square))
```

```python
plt.plot(x, f_x, label="MLE")
plt.legend()
plt.show()
```

# 랜덤추출 & MLE

```python
n = 5
mul = np.random.randint(-20, 20)
sample = (np.random.uniform(size=n)-0.5)*mul
print("Sample:", sample)

mean = np.mean(sample)
sigma_square = np.std(sample)**2
print("Mean:", mean)
print("Variance:", sigma_square)


x = np.linspace(-10, 10, 1000)
f_x = 1/(np.sqrt(2*np.pi*sigma_square))*np.exp(-(x-mean)**2/(2*sigma_square))

plt.plot(x, f_x, label="Estimation")
plt.legend()
plt.show()
```
