# 푸리에 변환

\begin{equation}
F(w) = ∫^∞_∞f(t)e^{-iwt}dt
\end{equation}


```python
import numpy as np
import matplotlib.pyplot as plt
```

# 신호
\begin{equation}
    f(t) = 10sin(2π*2*t) + 0.5sin(2π*25*t)
\end{equation}

```python
fs = 100
t = np.arange(0, 3, 1/fs)
frequency1 = 2
frequency2 = 25
signal = 10 * np.sin(2 * np.pi * frequency1 * t) + 0.5 * np.sin(2 * np.pi * frequency2 * t)

fft = np.fft.fft(signal) / len(signal)  
fft_magnitude = abs(fft)
f = np.linspace(-(fs / 2), fs / 2, len(t))
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2,1,1)
plt.plot(t,signal)
plt.grid()

plt.subplot(2,1,2)
plt.stem(f, np.fft.fftshift(fft_magnitude)) 
plt.grid()
```
